```python
import numpy as np
import pandas as pd                  
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('results.csv')
```

```python
data.head()
```

```python

def moving_avg(mA_range, data_input):   
    output = data_input.rolling(window = mA_range, center = False, on = "cat").mean().dropna()
    return output


mA_value = 150

chart_moving_avg = moving_avg(mA_value, data)   

plt.plot(chart_moving_avg['year'], chart_moving_avg['gat'], label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)")
plt.title("Global Average Temperature")
plt.show()
```

```python
plt.plot(chart_moving_avg['year'], chart_moving_avg['cat'], label='New Delhi')
plt.plot(chart_moving_avg['year'], chart_moving_avg['gat'], label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)")
plt.title("New Delhi Vs Global Average Temperature")
plt.show()
```
